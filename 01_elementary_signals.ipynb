Generate and plot:
- **Sine & Cosine:** `sin(Ï‰n)`, `cos(Ï‰n)`

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42) #FIXED random value
```

## ðŸ“š Unit Step Sequence

The **Unit Step Sequence** is a discrete-time signal defined as:

\[
u(n) =
\begin{cases} 
1 & \text{if } n \geq 0 \\
0 & \text{if } n < 0 
\end{cases}
\]

It is commonly used in signal processing and control systems to represent a signal that "turns on" at \( n = 0 \).


```python
# Generate Unit Step Sequence
n = np.arange(-10, 11, 1)  # Range from -10 to 10 increment 0.5
u_n = np.where(n >= 0, 1, 0)  # Unit step: 1 if n >= 0, else 0

# Plot Unit Step Sequence
plt.plot(n,u_n,label="Unit Step Sequence")
plt.scatter(n,u_n)
plt.title("Unit Step Sequence")
plt.xlabel("n")
plt.ylabel("u(n)")
plt.legend(loc='best')
plt.grid(True)
plt.show()
```


## ðŸ“š Ramp Sequence




```python
n=np.arange(-10,11,1)
ramp=np.where(n<0,0,n)

plt.scatter(n,ramp)
plt.plot(n,ramp,label="Ramp Sequence")
plt.xlabel("n")
plt.ylabel("r(n)")
plt.legend(loc='best')
plt.grid(True)
plt.show()

```

## Exponential Function

The **exponential function** is written as:
<math xmlns="http://www.w3.org/1998/Math/MathML" display="block"><semantics><mrow><mi>y</mi><mo stretchy="false">(</mo><mi>t</mi><mo stretchy="false">)</mo><mo>=</mo><mi>A</mi><mo>â‹…</mo><msup><mi>e</mi><mrow><mo stretchy="false">(</mo><mi>b</mi><mi>t</mi><mo stretchy="false">)</mo></mrow></msup></mrow><annotation encoding="application/x-tex">y(t) = A \cdot e^{(bt)}</annotation></semantics></math>
Where:

- <math xmlns="http://www.w3.org/1998/Math/MathML"><semantics><mrow><mi></mi></mrow><annotation encoding="application/x-tex"></annotation></semantics></math>A = Amplitude (scaling factor)
- <math xmlns="http://www.w3.org/1998/Math/MathML"><semantics><mrow><mi></mi></mrow><annotation encoding="application/x-tex"></annotation></semantics></math>b = Rate of growth or decay (positive for growth, negative for decay)
- <math xmlns="http://www.w3.org/1998/Math/MathML"><semantics><mrow><mi></mi></mrow><annotation encoding="application/x-tex"></annotation></semantics></math>t = Time (or independent variable)
- <math xmlns="http://www.w3.org/1998/Math/MathML"><semantics><mrow><mi></mi></mrow><annotation encoding="application/x-tex"></annotation></semantics></math>e = Euler's number (approx. 2.71828)

```python
# Parameters
A = 1      # Amplitude
b = 0.5    # Growth rate (decay for negative values)
t = np.linspace(0, 10, 500)  # Time from 0 to 10, 500 points

# Exponential function
y = A * np.exp(b * t)

# Plotting
plt.plot(t, y)
plt.title('Exponential Growth')
plt.xlabel('Time (t)')
plt.ylabel('y(t)')
plt.grid(True)
plt.show()
```

## ðŸ“š Sine Wave 

The sine wave is generated using the following formula:
### y(t) = A Ã— sin(2Ï€ft + Ï†)

Where:
- **A** = Amplitude
- **f** = Frequency (Hertz)
- **t** = Time (seconds)
- **Ï†** = Phase Shift (radians)
## ðŸ“š Cosine Wave 

The cosine wave is generated using the following formula:
### y(t) = A Ã— cos(2Ï€ft + Ï†)

Where:
- **A** = Amplitude
- **f** = Frequency (Hertz)
- **t** = Time (seconds)
- **Ï†** = Phase Shift (radians)



```python
x=np.linspace(0,2*np.pi,1000) # 1000 points between 0 and 2Ï€
y1=np.sin(x) #Sine Wave
y2=np.cos(x) #Cos Wave

fig = plt.figure(figsize=(10,4)) #Figure Size

ax1=fig.add_subplot(1,2,1) #
ax1.plot(x,y1,color='green' , label='Sine wave')
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.legend(loc='best')
plt.axhline(y=0,color='black') #horizontal line
plt.axvline(x=0,color='red') #vertical line
plt.title("Sine Wave")





ax2=fig.add_subplot(1,2,2)
ax2.plot(x,y2,color='green' , label="Cos Wave")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.legend(loc='best')
plt.axhline(y=0,color='black') #horizontal line
plt.axvline(x=0,color='red') #vertical line
plt.title("Cos Wave")
plt.show() #Render Plot



```
